```python
import numpy as np
```

```python
f = open("../data/casp7/training_30.txt")
```

```python
lines = f.readlines()
```

```python
def coords_splitter(lis, splice):
    coords = []
    for c in lis:
        coords.append([float(i) for i in c.split(splice)])
    return coords
```

```python
names = []
seqs = []
coords = []
pssms = []

for i in range(0, len(lines)):
    if len(coords) == 995:
        break
    
    if lines[i] == "[ID]\n":
        names.append(lines[i+1])
    elif lines[i] == "[PRIMARY]\n":
        seqs.append(lines[i+1])
    elif lines[i] == "[TERTIARY]\n":
        coords.append(coords_splitter(lines[i+1:i+4], "\t"))
    elif lines[i] == "[EVOLUTIONARY]\n":
        pssms.append(coords_splitter(lines[i+1:i+22], "\t"))
    
    if (i%100) == 0:
        print("{}/{} done".format(len(names), 950))
```

```python
def coord_normalize(coords):
    return (sum([c*c for c in coords]))**0.5
```

```python
print(len(seqs))
sel_prot_idxs = []
for i in range(0, len(seqs)):
    if len(seqs[i])< 200:
        sel_prot_idxs.append(i)
print(len(sel_prot_idxs))
```

```python


dists = []
for k in sel_prot_idxs:
    dist = []
    for i in range(len(coords[k][1])):
        if i%3 == 1:
            inter_aa_d = []
            for j in range(len(coords[k][1])):
                if j%3 == 1:
                    inter_aa_d.append(coord_normalize(np.array([coords[k][0][i],coords[k][1][i],coords[k][2][i]])-np.array([coords[k][0][j],coords[k][1][j],coords[k][2][j]])))
            dist.append(inter_aa_d)
    dists.append(dist)
    
    if len(dists)%100 == 0:
        print("{}/{}".format(len(dists), 500))
                    
                    
                    
```

```python
n = 1
print("id: ", names[n])
print("seq: ", seqs[n]) 
print("sample coord: ", coords[n][0][0]) 
print("sample dist: ", dists[n][0][4])
```

```python


with open("../Data/Processed/full_under_200.txt", "w") as f:
    ans = [0]
    for k in sel_prot_idxs:
        ans.append(ans[len(ans)-2] + 1)
        f.write("\n[ID]\n")
        f.write(names[k])
        f.write("\n[PRIMARY]\n")
        f.write(seqs[k])
        f.write("\n[EVOLUTIONARY]\n")
        f.write(str(pssms[k]))
        f.write("\n[TERTIARY]\n")
        f.write(str(coords[k]))
        f.write("\n[DIST]\n")
        if(len(dists[ans[len(ans) - 2]][0])> 200):
            print("Error")
            break
        else:
            f.write(str(dists[ans[len(ans)-1]]))
```
